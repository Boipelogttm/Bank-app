```python
class ATM:
    def __init__(self):
        self.username = "Themba"
        self.password = "admin123"
        self.login_attempts = 3
        self.balance = 2500
        

    def credentials(self):
        while self.login_attempts < 3:
            username = input("Please Enter your username: ")
            password = input("Please Enter your password: ")

            if username == self.username and password == self.password:
                print("Login successful.")
                self.login_attempts = 0
                self.display_menu()
                break
            else:
                print("Invalid username or password. Please try again.")
                self.login_attempts += 1
        else:
            print("Your account is locked.")
            exit()

    def display_menu(self):
        while True:
            print("\n***** Welcome to Sibula Banking App *******")
            print("1. Check Balance")
            print("2. Withdraw")
            print("3. Deposit")
            print("4. Display Users")
            print("5. Exit")

            choice = input("Please enter choice from options 1,2,3,4,5:")

            if choice == '1':
                self.check_balance()
            elif choice == '2':
                self.withdraw()
            elif choice == '3':
                self.deposit()
            elif choice == '4':
                self.display_users()
            elif choice == '5':
                print("Thank you for using Sibula Banking App. Goodbye!")
                exit()
            else:
                print("Invalid choice. Please enter valid choice .")
                

    def check_balance(self):
        print(f"Your balance is R{self.balance}")

    def withdraw(self):
        amount = float(input("Enter the amount to withdraw: "))
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"R{amount} withdrawn successfully.")
        else:
            print("You do not have the funds. Please try again.")

    def deposit(self):
        amount = float(input("Enter the amount to deposit: "))
        if amount > 0:
            self.balance += amount
            print(f"R{amount} deposited successfully.")
        else:
            print("Invalid deposit amount. Please try again.")

    def display_users(self):
        print("Approved Users:")
        for admin in self.administrators:
            print(admin)


if __name__ == "__main__":
    atm = ATM()
    atm.credentials()
```
